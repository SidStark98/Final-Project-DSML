```python
#Amazon Data
amzn_data = web.DataReader(name='AMZN', data_source='yahoo', start=start_date, end=end_date)
amzn_data
```

```python
#Data Cleaning
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
amzn_df = amzn_data
amzn_df.head()
```

```python
amzn_df.isnull().sum()
```

```python
amzn_df.dropna(how='all')
```

```python
amzn_df.isnull().any()
```

This shows that the data is clean.

```python
# Change in stock price over time.
Change_in_Price = amzn_df['High'] - amzn_df['Low']
Change_in_Price
```

```python
amzn_data['Price Change'] = Change_in_Price
amzn_data.head()
```

```python
import matplotlib.pyplot as plt
# Rolling Average for 20 Days
Rolling_Average_20 = amzn_df['Adj Close'].rolling(window=5).mean()
Rolling_Average_20.head(20)
```

```python
# Adding to dataframe
amzn_data['20 Day Rolling Avg'] = Rolling_Average_20
amzn_data.head(20)
```

```python
# Rolling average for first 50 days
Rolling_Average_50 = amzn_df['Adj Close'].rolling(window=5).mean()
Rolling_Average_50.head()
```

```python
# adding to dataframe
amzn_data['50 Day Rolling Avg'] = Rolling_Average_50
amzn_data.head()
```

```python
# Rolling average for first 100 days
Rolling_Average_100 = amzn_df['Adj Close'].rolling(window=5).mean()
Rolling_Average_100.head()
```

```python
# Adding to dataframe
amzn_data['100 Day Rolling Avg'] = Rolling_Average_100
amzn_data.head()
```

```python
# Daily Return Average
Daily_Return = amzn_df['Adj Close'].pct_change()
Daily_Return
```

```python
#Adding to dataframe
amzn_data['Daily Retrun Avg'] = Daily_Return
amzn_data.head()
```

```python
def trend(x):
  if x > -0.015 and x <= 0.015:
    return 'Slight or No change'
  elif x > 0.015 and x <= 0.04:
    return 'Slight Positive'
  elif x < -0.015 and x >= -0.4:
    return 'Slight Negative'
  elif x > 0.04 and x <= 0.06:
    return 'Positive'
  elif x < -0.04 and x >= -0.06:
    return 'Negative'
  elif x > 0.06 and x <= 0.07:
    return 'Among top gainers'
  elif x < -0.06 and x >= -0.07:
    return 'Among top losers'
  elif x > 0.07:
    return 'Bull run'
  elif x <= -0.07:
    return 'Bear drop'
```

```python
amzn_df['Trend'] = amzn_df['Daily Retrun Avg'].apply(lambda x: trend(x))
amzn_df.head()
```

```python
df1 = amzn_df
```
